```python
import cv2
import numpy as np
import matplotlib.pyplot as plt 

DESK_PATH = "./cv_desk.png"
COVR_PATH = "./cv_cover.jpg"

img_desk = cv2.imread(DESK_PATH)
img_covr = cv2.imread(COVR_PATH)

orb = cv2.ORB_create()
kp_desk = orb.detect(img_desk, None)
kp_desk_, des_desk = orb.compute(img_desk, kp_desk)

kp_covr = orb.detect(img_covr, None)
kp_covr_, des_covr = orb.compute(img_covr, kp_covr)
```

```python
bfs_matcher = cv2.BFMatcher()
matches = bfs_matcher.match(des_desk, des_covr)
```

```python
img_vis = cv2.drawMatches(
    img_desk,
    kp_desk,
    img_covr,
    kp_covr,
    matches[:10],
    None,
    flags=2
)

plt.imshow(img_vis)
plt.show()
```
